```python
# Checking available URL/PDF

# Define months in Spanish
import requests
months=['ENERO','FEBRERO','MARZO','ABRIL','MAYO','JUNIO','JULIO','AGOSTO','SEPTIEMBRE','OCTUBRE','NOVIEMBRE','DICIEMBRE']
#url = 'https://www.controlhidrocarburos.gob.ec/wp-content/uploads/Prod-Mensual-Fiscalizada-'+date+'.pdf'

working_url=[]
non_working_url=[]
#years = ['2018', '2019', '2020']

#for y in years:
for m in months:
    start_date=1
    stop_date=31

    while start_date <= stop_date:
        start_str=str(start_date)
        if len(start_str) == 1:
            start_str='0'+start_str
        else:
            start_str
        url = 'https://www.controlhidrocarburos.gob.ec/wp-content/uploads/Prod-Mensual-Fiscalizada-'+start_str+'-'+m+'-2020.pdf'

        response=requests.get(url)
        # Check for errors
        try:
            response.raise_for_status()
        except Exception as exc:
            #print('There was a proble:%s' % (exc))
            non_working_url.append(url)
        else:
            #print('This '+url+'is working')
            working_url.append(url)
        start_date=start_date+1
        

print('loop has finished')



# exporting all the working URL in a csv:
import pandas as pd
df_url=pd.DataFrame(working_url)
print(df_url)
csv_url=df_url.to_csv('csv_url.csv',index=False)
```
