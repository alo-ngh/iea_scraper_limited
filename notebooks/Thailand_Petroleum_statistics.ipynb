## Excel Files Download

```python
 # let's download data

import requests
from pathlib import Path

logger = logging.getLogger(__name__)
provider = 'th_gov_eppo'

download_dir = Path('.') / 'filestore'

urls = ['#REF',
        'http://www.eppo.go.th/epposite/images/Energy-Statistics/energyinformation/Energy_Statistics/Petroleum/T02_01_05.xls',
       'http://www.eppo.go.th/epposite/images/Energy-Statistics/energyinformation/Energy_Statistics/Petroleum/T02_02_02.xls',
       'http://www.eppo.go.th/epposite/images/Energy-Statistics/energyinformation/Energy_Statistics/Petroleum/T02_02_02-1.xls',
       'http://www.eppo.go.th/epposite/images/Energy-Statistics/energyinformation/Energy_Statistics/Petroleum/T02_03_02.xls',
       'http://www.eppo.go.th/epposite/images/Energy-Statistics/energyinformation/Energy_Statistics/Petroleum/T02_03_07.xls',
       'http://www.eppo.go.th/epposite/images/Energy-Statistics/energyinformation/Energy_Statistics/Petroleum/T02_03_09.xls',
       'http://www.eppo.go.th/epposite/images/Energy-Statistics/energyinformation/Energy_Statistics/Petroleum/T02_04_02.xls']

for url in urls:
    logger.info(f'Downloading {url}')
    response = requests.get(url)
    if response.ok:
        file = download_dir / f"{provider}_{url.split('/')[-1]}"
        logger.debug(f'Response OK: downloading to {file}')
        logger.info(f'Saving file to {file}')
        file.write_bytes(response.content)
```

## Read The File to a data frame

```python
import pandas as pd
from pathlib import Path

logger = logging.getLogger(__name__)

download_dir = Path('.') / 'filestore'
file = download_dir / 'th_gov_eppo_T02_01_04.xls'

try:
    logger.info(f"Opening {file}")
    #read sheet and convert it to data frame
    data = pd.read_excel(file,header=None, skiprows=3)
    # display whole sheet data
    display(data)
except Exception as e:
    raise e
```

### Transform the file

```python
## Remove annual, y-t-d and growth rate figures. Keep only monthly

# propagation non-null values forward for the first row
data.iloc[0] = data.iloc[0].fillna(method='ffill')


# find the max number of column which contains MONTH 
result = data.iloc[0].str.contains('MONTH')
columnNames = list(result[result == True].index)
print('Names of columns which contains MONTH:', columnNames)
max_col_name =max(columnNames)
print('max:',max_col_name)

# find the min number of column which contains Growth Rate 
result_1 = data.iloc[0].str.contains('GROWTH RATE')
columnNames_1 = list(result_1[result_1 == True].index)
print('Names of columns which contains GROWTH RATE:', columnNames_1)
min_col_name =min(columnNames_1)
print('min:',min_col_name)

#Selecting a subset of the dataframe based on the column's location
list_of_columns=[0]+list(range((max_col_name+1),min_col_name))
new_data = data.iloc[:,list_of_columns]
display(new_data)
```
