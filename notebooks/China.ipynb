# Fixing files for JAN-FEB 2020 extraction

## 1) Fix file encodings and delete tabs

The file 'cn_customs_crudeoil_imp_202002.csv' is encoded in 'iso-8859-1' instead of 'utf-8'.
This file has also \t as part of the separator.

The file 'cn_customs_oil_prod_202002.csv' is in 'iso-8859-1' when 'windows-1252' is expected and it has '\t' as part of the separator.

```python
import pandas as pd
from scraper.settings import FILE_STORE_PATH

files = ["cn_customs_crudeoil_imp_202002.csv", "cn_customs_oilprod_imp_202002.csv", "cn_customs_oilprod_exp_202002.csv"]
for i, file in enumerate(files):
    file_path = FILE_STORE_PATH / file
    if i == 0:
        print(f'{i} Fixing file {file_path}: converting from ISO-8859-1 to UTF-8 and removing tabs...')
        file_path.write_text(file_path.read_text(encoding='iso-8859-1').replace("\t", ""), encoding='utf-8')
    else:
        print(f'{i} Fixing file {file_path}: converting from ISO-8859-1 to windows-1252 and removing tabs...')
        file_path.write_text(file_path.read_text(encoding='iso-8859-1').replace("\t", ""), encoding='windows-1252')
```

## 2) Fix Date of data column

This column has as value '202001~02' when '202002' is expected.

```python
from scraper.settings import FILE_STORE_PATH


files = ["cn_customs_crudeoil_imp_202002.csv", "cn_customs_oilprod_imp_202002.csv", "cn_customs_oilprod_exp_202002.csv"]
for i, file in enumerate(files):
    file_path = FILE_STORE_PATH / file
    print(f'{i} Fixing file {file_path}')
    if i == 0:
        file_path.write_text(file_path.read_text(encoding='iso-8859-1').replace("202001~02", "202002"), encoding='utf-8')
    else:
        file_path.write_text(file_path.read_text(encoding='iso-8859-1').replace("202001~02", "202002"), encoding='windows-1252')
    
```
