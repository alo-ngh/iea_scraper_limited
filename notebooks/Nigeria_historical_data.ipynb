```python
# create the function for the trabsformation of 2008

def transform_data_2008():
    # read the data into the tabular
    production_year : str = '2008'
    df_2008 = tabula.read_pdf(file_link_2[0],pages = '23',output_format ='dataframe')
    # slice the dataframe
    final_df = df_2008[0]
    # drop all null values
    final_df = final_df.dropna(how='all')
    # move the columns into place
    final_df.columns = final_df.iloc[0].values
    # remove unwanted spaces
    final_df = final_df.iloc[3:,:]
    # drop row
    final_df.drop(labels=4,axis=0,inplace=True)
    
    # split the value into the column position 
    value_1 = final_df.iloc[4].values[12].split('00',1)[1]
    value_2 = final_df.iloc[4].values[12].split('00',1)[0] + '00'
    
    # assign the values into the appropriate position
    final_df.iloc[4].values[13] = value_1
    # assign to the total space
    final_df.iloc[4].values[12] = value_2
    # remove the pat of the dataframe not needed
    final_df = final_df.iloc[:35,:]
    # convert the column naEmes
    new_columes = ['entity','JANUARY','FEBRUARY','MARCH','APRIL','MAY','JUNE','JULY','AUGUST','SEPTEMBER','OCTOBER','NOVEMBER','DECEMBER','TOTAL','PERCENTAGE SHARE']
    final_df.columns = new_columes
    
    # drop the columns
    final_df.drop(labels=['TOTAL','PERCENTAGE SHARE'],axis=1,inplace=True)
    # rename a row
    final_df= final_df.rename(index={'(Condensate)':'Forcados condensate'})
    # remove the commas in the number
    final_df = final_df.replace(',','', regex=True)
    # reset the index
    final_df =final_df.reset_index(drop=True)
    # drop unwanted rows
    final_df.drop(labels =[22,30,31,32],axis=0,inplace=True)
    # reset the index
    final_df =final_df.reset_index(drop=True)
    # create the product column
    final_df['product'] = 'CRUDEOIL'
    final_df['entity'][9] = final_df['entity'][9].replace('- ','-')
    # arrage data
    # combine the columns
    final_df.iloc[0] = final_df.iloc[0][1:-1].astype(float)+ final_df.iloc[1][1:-1].astype(float)
    final_df.iloc[0][-1] = 'CRUDEOIL'
    final_df.iloc[0][0] = 'Forcados'
    final_df.iloc[9] = final_df.iloc[9][1:-1].astype(float)+ final_df.iloc[23][1:-1].astype(float) + final_df.iloc[24][1:-1].astype(float) + final_df.iloc[25][1:-1].astype(float)
    final_df.iloc[9][-1] = 'CRUDEOIL'
    final_df.iloc[9][0] = 'QUA_IBOE'
    final_df.iloc[7] = final_df.iloc[7][1:-1].astype(float)+ final_df.iloc[26][1:-1].astype(float)
    final_df.iloc[7][-1] = 'CRUDEOIL'
    final_df.iloc[7][0] = 'AMENAM'
    final_df.iloc[4] = final_df.iloc[4][1:-1].astype(float)+ final_df.iloc[27][1:-1].astype(float)
    final_df.iloc[4][-1] = 'CRUDEOIL'
    final_df.iloc[4][0] = 'ESCRAVOS'
    final_df.iloc[10] = final_df.iloc[10][1:-1].astype(float)+ final_df.iloc[28][1:-1].astype(float)
    final_df.iloc[10][-1] = 'CRUDEOIL'
    final_df.iloc[10][0] = 'OSO'
    # retain only the first name of the fields
    final_df['entity'] =final_df['entity'].map(lambda x : x.split(' ')[0].replace('-','_').upper())
    # drop row 
    final_df.drop(labels=[1,23,24,25,26,27,28],axis =0,inplace=True)
    # change the name
    final_df.iloc[21][0] = 'SEA EAGLE'
    # reset the name
    final_df =final_df.reset_index(drop=True)
    
    # melt the datframe into the required format
    final_df = pd.melt(final_df,id_vars=['entity','product'],var_name='period',value_name='value')
    
    # change the period format
    final_df['period'] = final_df['period'].str[:3] + str(production_year)
    
    return final_df   
```
