```python
from bs4 import BeautifulSoup
import urllib.request
import pandas as pd
import numpy as np
import datetime
import sqlalchemy

def find_link_address():
    url = 'https://www.gov.uk/government/statistics/oil-and-oil-products-section-3-energy-trends'
    page = urllib.request.urlopen(url)
    soup = BeautifulSoup(page, 'html.parser')
    a = soup.find(lambda tag:tag.name=="a" and "road fuel sales and stock levels (weekly data) - ODS" in tag.text)
    link = a['href']
    return link

def download_data(link):
    df = pd.read_excel(link, sheet_name='Data', skiprows=6, engine='odf')
    df['Date'] = pd.to_datetime(df['Date'], format='%Y-%m-%d')
    df.rename(columns={'Date':'date', 
                       'Weekday':'weekday', 
                       'Fuel Type':'fuel_type', 
                       'Region':'region',
                       'Average Sales Per Filling Station (litres)':'sales_litres'}, inplace=True)
    df.set_index('date', inplace=True)
    return df

link = find_link_address()
df = download_data(link)
df
```
